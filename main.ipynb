```python
import matplotlib.pyplot as plt

def graficar_resultados(porcentajes):
    modelos = ['KNN', 'SVM', 'Decision Tree']

    plt.bar(modelos, porcentajes, color=['blue', 'green', 'red'])
    plt.xlabel('Modelos de Clasificación')
    plt.ylabel('Porcentaje de Exactitud')
    plt.ylim(0, 100)  # Establecer el rango del eje y de 0 a 100 para representar porcentajes
    
    # Agregar etiquetas a las barras
    for i, porcentaje in enumerate(porcentajes):
        plt.text(i, porcentaje + 1, f'{porcentaje:.2f}%', ha='center', va='bottom')

    plt.show()

# Ejemplo de uso:
porcentajes = [91.47, 90.34, 82.31]  # Reemplazar con tus propios porcentajes
graficar_resultados(porcentajes)
```

```python
import matplotlib.pyplot as plt

def graficar_resultados_deep_learning(ann_porcentaje, cnn_porcentaje):
    modelos = ['ANN', 'CNN']

    plt.bar(modelos, [ann_porcentaje, cnn_porcentaje], color=['purple', 'orange'])
    plt.xlabel('Modelos de Deep Learning')
    plt.ylabel('Precisión')
    plt.ylim(0, 100)  # Establecer el rango del eje y de 0 a 100 para representar porcentajes

    # Agregar los porcentajes en la parte superior de las barras
    for i, porcentaje in enumerate([ann_porcentaje, cnn_porcentaje]):
        plt.text(i, porcentaje + 1, f'{porcentaje:.2f}%', ha='center', va='bottom')

    plt.show()

# Ejemplo de uso:
ann_porcentaje = 88.49  # Reemplazar con tu propio porcentaje de precisión para ANN
cnn_porcentaje = 98.89  # Reemplazar con tu propio porcentaje de precisión para CNN

graficar_resultados_deep_learning(ann_porcentaje, cnn_porcentaje)

```
